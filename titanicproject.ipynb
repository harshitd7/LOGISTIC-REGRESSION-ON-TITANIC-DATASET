```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("training_titanic_x_y_train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(["Name","Ticket"],axis=1,inplace=True)
```

```python
df.head()
```

```python
def male_ohe(g):
    if g == "male":
        return 1
    else:
        return 0

def female_ohe(g):
    if g == "female":
        return 1
    else:
        return 0
    

df["Male"] = df["Sex"].apply(male_ohe)
df["Female"] = df["Sex"].apply(female_ohe)
```

```python
df.drop("Sex",axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
def cabin(x):
    if x is np.nan:
        return 0
    else:
        return 1
df["Cabin"]=df.Cabin.apply(cabin)
```

```python
df.head(20)
```

```python
def cherbourg(s):
    if s=="C":
        return 1
    else:
        return 0
    
def queenstown(s):
    if s=="Q":
        return 1
    else:
        return 0

def southamptom(s):
    if s=="S":
        return 1
    else:
        return 0
    
df["Cherbourg"] = df.Embarked.apply(cherbourg)
df["Queenstown"] = df.Embarked.apply(queenstown)
df["Southamptom"] = df.Embarked.apply(southamptom)
df.drop("Embarked",axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
df.head()
```

```python
df.columns
```

```python
Y=df['Survived']
del df['Survived']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df,Y)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
